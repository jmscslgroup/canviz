In this notebook, we will write a code to gather velocity data from all the drives we have done so far.

```python
from strym import strymread
import strym
import glob
import pandas as pd
import os
```

```python
parentfolder = "/home/ivory/CyverseData/JmscslgroupData/PandaData/"
dbcfile = '/home/ivory/VersionControl/Jmscslgroup/strym/examples/newToyotacode.dbc'
csvlist = []
folderlist = glob.glob(parentfolder+"*")
speedlist = []
for datafolder in folderlist:
    csvlisttmp = glob.glob(datafolder+"/*.csv")
    for f in csvlisttmp:
        if "GPS" in f:
            continue
        csvlist.append(f)
```

```python
vel_directory = '/home/ivory/CyverseData/JmscslgroupData/speed/'
if not os.path.exists(vel_directory):
    os.mkdir(vel_directory)
for csv in csvlist:
    print("\nReading the CSV file {}".format(csv))
    pathsplit = csv.split('/')
    # print(pathsplit[-1])
    # Proceed only when csvfile is not empty
    speedfile = vel_directory + pathsplit[-1][0:-4] + "_speed.csv"
    if os.path.exists(speedfile):
        continue
    
    r = strymread(csvfile=csv, dbcfile=dbcfile)
    if r.dataframe.shape[0] != 0 :
        speed = r.speed()
        print("Writing Speed Data to CSV file {}\n".format(speedfile))
        speed.to_csv(speedfile)
```
