# Data throughout comparison betwene C++ code and Python for Toyota RAV4 CAN Data
We compare the difference of data quality of CAN data obtained through live recording using C++ API libpanda and using pythonic implementation. There are two sets of data we obtained, one through libpanda written in C++ and through python.

```python
from strym import strymread
import strym
import matplotlib.pyplot as plt
import numpy as np

msgsOfInterest = [180,37,384]
```

## Reading data captured using C Code

```python
cdata = '../../PandaData/2020_03_09/2020-03-09-18-18-29_CAN_Messages.csv'
import pandas as pd
```

```python
cframe = pd.read_csv(cdata)
```

```python
cframe
```

```python
csorted = cframe.sort_values(by='MessageID')
```

```python
csorted
```

```python
cnt = csorted['MessageID'].value_counts()
```

```python
cnt = cnt.sort_index(ascending=True)
```

```python
cnt
```

## Plot the count of messages

```python
plt.style.use('ggplot')
plt.rcParams["figure.figsize"] = (16,8)
fig, ax = plt.subplots(1,1)
fig.tight_layout(pad=5.0)
cnt.plot(kind='bar', ax=ax)
ax.tick_params(axis="x")
ax.tick_params(axis="y")
```

## Frequency Analysis of messages

```python
for index in msgsOfInterest:
    m_data = cframe[cframe['MessageID'] == index ]
    n_msg = m_data.shape[0]
    percent = 40
    n_msg_to_trunc = (n_msg*percent)/100
    n_msg_to_trunc= int(n_msg_to_trunc)
    print(n_msg_to_trunc)
    m_data = m_data[n_msg_to_trunc:]

    strymread.ranalyze(m_data, title ="Message ID: " + str(index), savefig = False)
```

## Reading data captured using Py Code

```python
pydata = '../../PandaData/2020_03_06/2020-03-06-11-51-50-261530__CAN_Messages.csv'
pyframe = pd.read_csv(pydata)
```

```python
pysorted = pyframe.sort_values(by='MessageID')
pycnt = pysorted['MessageID'].value_counts()
pycnt = pycnt.sort_index(ascending=True)
```

```python
plt.style.use('ggplot')
plt.rcParams["figure.figsize"] = (16,8)
fig, ax = plt.subplots(1,1)
fig.tight_layout(pad=5.0)
pycnt.plot(kind='bar', ax=ax)
ax.tick_params(axis="x")
ax.tick_params(axis="y")
```

```python
for index in msgsOfInterest: #pycnt.index:
    m_data = pyframe[pyframe['MessageID'] == index ]
    strymread.ranalyze(m_data, title ="Message ID: " + str(index), savefig = False)
```

## Side by side comparison by data throughput    
    

```python
for index in msgsOfInterest:
    m_data = pyframe[pyframe['MessageID'] == index ]
    strymread.ranalyze(m_data, title ="MessageID " + str(index) + " Using Python", savefig=True)
    
    n_data = cframe[cframe['MessageID'] == index ]
    n_data = cframe[cframe['MessageID'] == index ]
    n_msg = n_data.shape[0]
    percent = 40
    n_msg_to_trunc = (n_msg*percent)/100
    n_msg_to_trunc= int(n_msg_to_trunc)
    print(n_msg_to_trunc)
    n_data = n_data[n_msg_to_trunc:]
    strymread.ranalyze(n_data, title ="MessageID " + str(index) + " Using C", savefig = True)
```

Clearly, data captured through C++ API has better data rate and it dropped less packets than one obtained through Pythonic implementation
